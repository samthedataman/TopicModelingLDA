##Import Packages

```
import pyLDAvis
import sklearn
import pandas as pd
from sklearn import datasets
from sklearn import feature_extraction
from sklearn import decomposition
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import re
import numpy as np
import nltk
import string

nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

##Fetch Data

```
data = '/content/drive/My Drive/ReviewsFinal2.csv'
df = pd.read_csv(data)
df = df.dropna(0)
#h = "Construction"
#Construction  = df[df['MARKET'] == h]
#clean text prior to submitting to model
def cleanText(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text
#df = df[(df['INDUSTRY'] == 'Internet')]
#df = df.where(df['OVERALL_RATING'] <= 3)
docs_raw = (df['CONS_TEXT'].to_list())
docs_raw = [x for x in docs_raw if not isinstance(x, int)]
docs_raw = [x for x in docs_raw if not isinstance(x, float)]
df
```

```

docs_raw = (df['PROS_TEXT'].to_list())
docs_raw = [x for x in docs_raw if not isinstance(x, float)]
docs_raw
```

##Convert text to document term frequency Matrix
Raw docs are converted into document term matrix

```
tf_vectorizer = CountVectorizer(strip_accents='unicode',
                                stop_words = 'english',
                                lowercase =True,
                                token_pattern = r'\b[a-zA-Z]{4,}\b',
                                max_df =.5,
                                min_df = 10)
dtm_tf = tf_vectorizer.fit_transform(raw_documents=docs_raw)
```

##Generate and Fit Model with 10 topics

```
lda_tf=LatentDirichletAllocation(n_components=10
                                 ,random_state=0)
lda_tf.fit(dtm_tf)
```

##Visulize Text Data

```
pyLDAvis.sklearn.prepare(lda_tf,dtm_tf,tf_vectorizer)
```
